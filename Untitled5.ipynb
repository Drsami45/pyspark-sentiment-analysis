```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, split, explode, lower, regexp_replace

# Start Spark Session
spark = SparkSession.builder \
    .appName("SocialMediaSentimentAnalysis") \
    .getOrCreate()
```

```python
df = spark.read.csv(r"C:\Users\User\Downloads\archive (4)\training.1600000.processed.noemoticon.csv",header=False, inferSchema=True)
df = df.select(col("_c0").alias("target"), col("_c5").alias("text"))
df.show(5)
```

```python
df.groupBy("target").count().show()
```

```python
negative_df = df.filter(col("target") == 0)
positive_df = df.filter(col("target") == 4)

print(f"Total Negative Posts: {negative_df.count()}")
print(f"Total Positive Posts: {positive_df.count()}")
```

```python
top_keywords = df.withColumn("text", lower(col("text"))) \
    .withColumn("text", regexp_replace(col("text"), "[^a-z\\s]", "")) \
    .select(explode(split(col("text"), "\\s+")).alias("word")) \
    .filter(col("word") != "") \
    .groupBy("word") \
    .count() \
    .orderBy("count", ascending=False)

# Show the top 20 keywords
top_keywords.show(20)
```
